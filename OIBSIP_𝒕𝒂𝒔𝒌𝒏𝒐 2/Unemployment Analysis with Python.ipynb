# **Exploratory Data Analysis(EDA) or Data Visualisation**

### <b>Plotting a BarPlot for Regional Distribution of Unemployment </b>

```python
region_counts = df['Region'].value_counts()
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=160, 
                                   colors=sns.color_palette('viridis', len(region_counts)), wedgeprops=dict(width=0.4))
plt.title('Region Distribution Breakdown', fontsize=14)
for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_fontsize(10)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.tight_layout()
plt.show()
```

### <b>Plotting a Bubble Chart for Average Unemployment Rate by Region </b>

```python
unemployment_data = df[['Estimated Unemployment Rate (%)', 'Region']]
average_unemployment_by_region = unemployment_data.groupby('Region').mean().reset_index().sort_values('Estimated Unemployment Rate (%)')
fig = px.scatter(average_unemployment_by_region, 
                 x='Estimated Unemployment Rate (%)', 
                 y='Region', 
                 size='Estimated Unemployment Rate (%)', 
                 color='Region', 
                 title='Average Unemployment Rate by Region',
                 hover_name='Region', 
                 size_max=60)
fig.update_layout(xaxis_title='Average Unemployment Rate (%)', 
                  yaxis_title='Region', 
                  showlegend=False, 
                  height=600)
fig.show()
```

### <b>Plotting a Bar Chart for Average Unemployment Rates by Region and Area</b>

```python
average_unemployment = df.groupby(['Region', 'Area'])['Estimated Unemployment Rate (%)'].mean().reset_index()
fig = px.bar(average_unemployment, x='Region', y='Estimated Unemployment Rate (%)', 
             color='Area', 
             title='Average Unemployment Rates by Region and Area',
             text='Estimated Unemployment Rate (%)', 
            )
fig.update_layout(xaxis={'categoryorder': 'total descending'},
                  yaxis_title='Average Unemployment Rate (%)',
                  xaxis_title='Region',
                  height=600)
fig.show()
```

### <b> Plotting a Bar Plot with Grouping between Average Unemployment and Labour Participation Rates by Region </b>

```python
avg_rates = df.groupby('Region').agg({
    'Estimated Unemployment Rate (%)': 'mean',
    'Estimated Labour Participation Rate (%)': 'mean'
}).reset_index()

bar_fig = px.bar(avg_rates, 
                  x='Estimated Unemployment Rate (%)', 
                  y='Region', 
                  color='Estimated Labour Participation Rate (%)',
                  title='Average Unemployment and Labour Participation Rates by Region')
bar_fig.update_layout(
    xaxis_title='Average Unemployment Rate (%)',
    yaxis_title='Region',
    height=600,
    template='plotly_white'
)
bar_fig.show()
```
